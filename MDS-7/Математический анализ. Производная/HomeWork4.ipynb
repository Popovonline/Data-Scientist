### Уровень 0
посчитайте значение 1-ой и 2-ой производной фунции
f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd
%matplotlib inline
```

```python
from scipy.misc import derivative
from math import sin, cos
def f(x):
    return np.power(x,5) + 4 * sin(2*x) + cos(3*x + 3)

print ('Значение 1-ой производной функции в точке 1 равна {}'.format(derivative(f, 1.0, n=1, order=15)))
print ('значение 2-ой производной функции в точке 1 равна {}'.format(derivative(f, 1.0, n=2, order=15)))
```

### Уровень 1**
Постройте график первой и второй производной функции
(sin(2x+1))^5
на отрезке [-5; 5]

```python
# Графики построил двумя способами, в конце есть найденные полезные ссылки.
from sympy import diff, Symbol, cos, sin 
x = Symbol('x') #x, y = symbols('x y')
dfdx = diff(np.power(sin(2 * x + 1), 5))
dfdx
```

```python
d2fdx2 = diff(dfdx)
d2fdx2
```

```python
type(dfdx)
```

```python
type(print(dfdx))
```

```python
# Построим график первой производной:
x = np.linspace(-5.0, 5.0, 100)
y = [10 * sin(2*x[i] + 1) * 4 * cos(2*x[i] + 1) for i in range(len(x))]
plt.plot(x, y)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.title('График первой производной')
plt.ylabel('df\\dx')
plt.xlabel('x')

plt.show()
```

```python
# Построим график второй производной:
x = np.linspace(-5.0, 5.0, 100)
y = [-20 * sin(2*x[i] + 1) ** 5 + 80 * sin (2*x[i] + 1) ** 3 * 3 * cos(2*x[i] + 1) ** 2 for i in range(len(x))]
plt.plot(x, y)
plt.axvline(0, color='r')
plt.axhline(0, color='r')
plt.title('График второй производной')
plt.ylabel('$d^2f$\\d$x^2$')
plt.xlabel('x')

plt.show()
```

### Второй способ построения графиков производных:

```python
from scipy.misc import derivative
def f(x):
    return np.power(sin(2 * x + 1), 5)
derivative(f, 1.0, dx=1e-6)
```

```python
x = np.linspace(-5.0, 5.0, 100)
fx = [f(x[i]) for i in range(len(x))] #fx = f(x)
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
f2x = [derivative(f, xxx, dx=1e-6, n=2) for xxx in x]
```

```python
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative')
plt.plot(x, f2x, label='second derivative')
plt.legend()
plt.show()
```
